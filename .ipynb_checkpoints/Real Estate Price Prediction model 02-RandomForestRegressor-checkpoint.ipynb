```python
df_test.info()
```

```python
df_train=pd.read_csv(ds_path_train)
df_train
```

```python
price=df_train.Price
price
```

```python
fiature_names=df_train.columns.usecols = ['Id', 'DistrictId', 'Rooms', 'Square', 'LifeSquare',
       'KitchenSquare', 'Floor', 'HouseFloor', 'HouseYear', 'Ecology_1',
       'Social_1', 'Social_2', 'Social_3', 'Healthcare_1', 'Helthcare_2',
       'Shops_1', 'A', 'B', 'A.1', 'B.1', 'A.2', 'B.2']
fiature_names
```

```python
x = pd.DataFrame(df_train, columns=fiature_names)
x
```

```python
x2 = pd.DataFrame(df_test, columns=fiature_names)
x2
```

```python
y=pd.DataFrame(price, columns=['Price'])
y.info()
```

```python
# # del x['Price']
# x.dropna(axis='columns',how='all', inplace=True)
```

```python
# разбить на тестовую и обучающую выборки
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42) # shuffle=True,
```

```python
# print(x_train)
# print(x_test)
print(y_train)
# print(y_test)
```

```python
# линейная регрессия
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# переобучение
lr.fit(x_train, y_train)
```

```python
# получить предсказанные значения на объектах X_test с помощью метода .predict:
y_pred = lr.predict(x_test)
y_pred.shape
```

```python
# Создадим таблицу DataFrame чтобы сопоставить реальные значения с предсказанными. 
# Поскольку массив y_pred является двумерным, переведём его в одномерный, используя метод .flatten.
check_test = pd.DataFrame({
    "y_test": y_test["Price"],
    "y_pred": y_pred.flatten(),
})
check_test
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(check_test["y_pred"], check_test["y_test"]))

from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(check_test["y_pred"], check_test["y_test"]))

from sklearn.metrics import r2_score as r2
r2(y_test, y_pred) # sample_weight=None, multioutput='uniform_average'
```

```python
# предсказываем данные на тесте!!!
y_pred_t = lr.predict(x2)
y_pred_t
```

```python
# готовим ответ в виде для каггла
kaggle = pd.DataFrame({
    "Id": x2["Id"],
    "Price": y_pred_t.flatten(),
})
kaggle.info()
```

```python
# сохранить обработанные данные в csv
# print(df.to_csv(sep='\t', index=False))
kaggle.to_csv('kaggle_01.csv', index=False)
```

```python
# переобучение и оценка модели
from sklearn.ensemble import GradientBoostingRegressor
final_model = GradientBoostingRegressor(n_estimators=100, max_depth=7, random_state=42)
final_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score as r2, mean_absolute_error as mae, mean_squared_error as mse
import seaborn as sns
def evaluate_preds(true_values, pred_values):
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
          "MAE:\t" + str(round(mae(true_values, pred_values), 3)) + "\n" +
          "MSE:\t" + str(round(mse(true_values, pred_values), 3)))
    
    plt.figure(figsize=(10,10)) 
    
    sns.scatterplot(x=pred_values, y=true_values)
    
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    plt.show()
```

```python
y_train_preds = final_model.predict(x_train)
evaluate_preds(y_train, y_train_preds)
```
